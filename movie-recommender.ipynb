```python
import pandas as pd
import numpy as np
movies= pd.read_csv("tmdb_5000_movies.csv")
credits= pd.read_csv("tmdb_5000_credits.csv")
movies.head()
```

```python
credits.head()
```

```python
# information of movies dataset
print(movies.info())
print(movies.describe())
print(movies.columns)
movies['release_date']=pd.to_datetime(movies['release_date'])
movies['year']=movies['release_date'].dt.year
```

```python
#information of credits dataset
print(credits.info())
print(credits.describe())
print(credits.columns)
```

```python
movies=movies.merge(credits,on='title') #merging movies dataset with credits dataset on title column
movies.columns
movies.rename(columns={'movie_id_x':'movie_id'},inplace=True)
```

```python
#colums to consider
#genres, id, keywords, title, overview, cast, crew
movies=movies[['movie_id','title','overview','genres','keywords','cast', 'crew', 'year']]
```

```python
movies.head()
movies['year']=movies['year'].astype(str).str[:4] #converting year to string and taking only first 4 characters
```

```python
movies.isnull().sum()
movies.duplicated().sum()
```

```python
movies.dropna(inplace=True) #dropping null values
```

```python
movies.iloc[3].genres
```

```python
import ast
ast.literal_eval(movies.iloc[0].genres)
```

```python
#extracting only the genre names from the genres column and appending it to a list
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
movies['genres'].apply(convert) #for every movie, getting a list of genres  of that movie
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.iloc[0].cast
# taking names of first 4 actors from each movie cast
def convert_cast(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=4:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
movies['cast'].head()
```

```python
movies.iloc[0].crew

#extracting the director name and music composer name from crew column
def convert_crew(obj):
    L_dir=[]
    L_music=[]
    for i in ast.literal_eval(obj):
        #checking director field exists and extracting director name
        if i['job'] and i['job']=='Director':
            L_dir.append(i['name'])      
        #checking music composer field exists and extracting music composer                                
        if i['job'] and i['job']=='Original Music Composer':
            L_music.append(i['name'])
    return L_dir,L_music


```

```python
movies['crew']=movies['crew'].apply(convert_crew)
movies['crew'].head()
```

```python
#splitting the contents of overview column into list of words
movies['overview']=movies['overview'].apply(lambda x:x.split()) 
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
#creating new columns for director and music composer
movies['director']=movies['crew'].apply(lambda x:x[0]) # director -> first element of crew list
movies['music_composer']=movies['crew'].apply(lambda x:x[1]) #music composer -> second element of crew list

#removing spaces from director and music composer names
movies['director']=movies['director'].apply(lambda x:[i.replace(" ","") for i in x])
movies['music_composer']=movies['music_composer'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
#creating tags column by combining overview, genres, keywords, cast, director and music composer
movies['tags']= (movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['director']+movies['music_composer'])
movies.head()
```

```python
moviesdf=movies[['movie_id','title','tags']] #creating a new dataframe with only required columns   
moviesdf.head()
moviesdf['tags']=moviesdf['tags'].apply(lambda x:" ".join(x)) #joining the list of words to form a single string
```

```python
print(moviesdf.head())
moviesdf['tags']=moviesdf['tags'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
 #stemming the words -> love, loving, loved -> love
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
moviesdf['tags']=moviesdf['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv5=CountVectorizer(max_features=5000,stop_words='english')
vectors5=cv5.fit_transform(moviesdf['tags']).toarray() #creating vectors for every movie based on the tags column based upon 5000 most frequent words
cv2=CountVectorizer(max_features=2000,stop_words='english')
vectors2=cv2.fit_transform(moviesdf['tags']).toarray() #creating vectors for every movie based on the tags column based upon 2000 most frequent words
```

```python
#shape of the 2 vectors formed
print(vectors5.shape)
print(vectors2.shape)
```

```python
print(cv5.get_feature_names_out())
print(cv2.get_feature_names_out())
#alphabetically sorted feature names of both vectorizers
```

```python
#finding cosine distance between different vectors, less distance means more similar
from sklearn.metrics.pairwise import cosine_similarity
similarity5=cosine_similarity(vectors5)
similarity2=cosine_similarity(vectors2)
print(similarity5.shape)
print(similarity2.shape)
```

```python
def recommend(movie):
    movie_index=moviesdf[moviesdf['title']== movie].index[0]
    distances5=similarity5[movie_index]
    distances2=similarity2[movie_index]
    #sorting based upon similarity scores by converting distances into a list of tuples
    lis5=sorted(list(enumerate(distances5)), reverse=True, key=lambda x:x[1])[1:11]  #top 10 movies with highest similarity scores based upon 5000 features
    lis2=sorted(list(enumerate(distances2)), reverse=True, key=lambda x:x[1])[1:11] #top 10 movies with highest similarity scores based upon 2000 features

    print("Top 10 recommendations based upon 5000 features are: \n")
    for i in lis5:
        print(moviesdf.iloc[i[0]].title, movies.iloc[i[0]].year)
        #print(movies.iloc[i[0]].year)

    print("\n Top 10 recommendations based upon 2000 features are: \n")
    for i in lis2:
        #print(moviesdf.iloc[i[0]].title)
        #print(movies.iloc[i[0]].year)
        print(moviesdf.iloc[i[0]].title, movies.iloc[i[0]].year)
```

```python
#recommendations for a particular movie

recommend('Batman Begins')
```

```python
recommend('Spectre')
```

```python
recommend('Dr. No')
```
